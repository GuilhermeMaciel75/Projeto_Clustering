```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors 
```


## Ajustando dataset

```python
# Carregando o conjunto de dados
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
df = pd.read_csv(url, header=None)

# Obtendo os rótulos de classe
targets = df[0]

# Descartando a coluna de rótulo
df = df.drop(columns=[0])

# Convertendo o dataframe para uma matriz NumPy e atribui o tipo float aos dados
data = df.values
X = data.astype(float)
```

## Normalizando dataset e aplicando PCA

```python
# Normalizando os recursos usando o método Z-score
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Converter os recursos normalizados de volta para DataFrame
df_normalized = pd.DataFrame(X)

# Criando uma instância do modelo PCA com três componentes
pca = PCA(n_components=3)

# Ajustando o modelo aos dados
X_pca = pca.fit_transform(X)

# Tranformando o modelo de dados PCA em um dataframe em pandas
df_pca = pd.DataFrame(X_pca)
```

## Calculando a distância entre um ponto e seus vizinhos usando o método NearestNeighbors
### Encontrando o número ideal de eps



```python
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(df_pca)
distances, indices = nbrs.kneighbors(df_pca)

# Criando uma instância da classe NearestNeighbors 
# (encontra os dois vizinhos mais próximos de cada ponto no conjunto de dados)
neigh = NearestNeighbors(n_neighbors=2)

# # Ajustando o modelo ao conjunto de dados
nbrs = neigh.fit(df_pca)

# Aplicação do modelo treinado na base de dados df_pca, utilizando o método kneighbors()
# (encontra as distâncias e os índices dos dois vizinhos mais próximos de cada ponto em df_pca).
distances, indices = nbrs.kneighbors(df_pca)

```

```python
# Ordenando as distancias
distances = np.sort(distances, axis=0)
distances = distances[:,1]

# Plotando o resultado
plt.plot(distances)
```

O ponto de curvatura máxima esta entre 0.7 a 1

## Aplicando DBSCAN e encontrando o número de clusters e ruídos



```python
def dbs_full(eps_inicial, eps_final, acrescento):
  lista_eps = np.arange(eps_inicial, eps_final + acrescento, acrescento).tolist()

  for eps in lista_eps:
      # Criando uma instância do modelo DBSCAN com os parametros selecionados
      dbscan = DBSCAN(eps=eps, min_samples=5)

      # Ajustando o modelo aos dados
      dbscan.fit(df_pca)

      # Guardando as etiquetas de classe atribuídas a cada amostra
      labels = dbscan.labels_

      # Definindo e imprimindo o número de clusters e o número de amostras de ruído
      n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
      n_noise = list(labels).count(-1)
      print('eps:', eps)
      print('Número de clusters:', n_clusters)
      print('Número de amostras de ruído:', n_noise)

      # Criando uma figura 3D
      fig = plt.figure(figsize = (8, 8))
      ax = fig.add_subplot(111, projection='3d')

      # Plotando os dados em um gráfico de dispersão 3D
      ax.scatter(df_pca[0], df_pca[1], df_pca[2], c=labels, cmap='viridis')

      # Definindo rótulos de eixo e título do gráfico
      ax.set_xlabel('Recurso 1')
      ax.set_ylabel('Recurso 2')
      ax.set_zlabel('Recurso 3')
      ax.set_title('Resultado do DBSCAN em 3D')

      # Exibindo gráfico
      plt.show()
```

```python
dbs_full(0.7, 1, 0.05)
```

Analisando os resultados, o número ideal de eps está entre 0.8 e 0.85, podemos analisar ainda mais precisamente:

```python
dbs_full(0.8, 0.9, 0.01)
```
